```python
# Data Preprocessing
#converting categorical to numerical
from sklearn.preprocessing import LabelEncoder
label_encoder={}
for column in data.select_dtypes(include=['object']).columns:
    label_encoder[column]=LabelEncoder()
    data[column] = label_encoder[column].fit_transform(data[column])
data
```

```python
# Convert all float values to integers
data = data.applymap(lambda x: int(round(x)) if isinstance(x, float) else x)
data
```

```python
#spliting the data
x=data.iloc[:,:-1]
y=data.iloc[:,-1]
x,y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,train_size=0.8,random_state=42)
x_train,y_train
```

```python
# Standardize features by removing the mean and scaling to unit variance
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
x_train = scaler.fit_transform(x_train)
x_test = scaler.transform(x_test)
x_train,x_test
```

```python
#SVM Implementation
from sklearn import svm
svm_model=svm.SVC(kernel='linear')
```

```python
svm_model.fit(x_train,y_train)
```

```python
from sklearn.metrics import classification_report,accuracy_score,confusion_matrix
# Predict on the test data using the trained SVM model
y_pred_test = svm_model.predict(x_test)
cf=classification_report(y_pred_test,y_test)
print('classification_report',cf)
```

```python
accuracy = accuracy_score(y_test, y_pred_test)
print("Accuracy:", accuracy)
```

```python
cm=confusion_matrix(y_test, y_pred_test)
print("Confusion Matrix:",cm)
```

```python
#Visualization of SVM Results
plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt="d", cmap="Blues", cbar=False)
plt.xlabel('Predicted Labels')
plt.ylabel('True Labels')
plt.title('Confusion Matrix')
plt.show()
```

```python
#Comparison and Analysis
from sklearn.svm import SVC
kernels = ['linear', 'poly', 'rbf']
for kernel in kernels:
    svm_model = SVC(kernel=kernel)
    svm_model.fit(x_train, y_train)
    y_pred = svm_model.predict(x_test)
    accuracy = accuracy_score(y_test, y_pred)
    print("Kernel:", kernel)
    print("Accuracy:", accuracy)
    print(classification_report(y_test, y_pred))
    print("Confusion Matrix:")
    print(confusion_matrix(y_test, y_pred))
```

```python
# Parameter Tuning and Optimization
from sklearn.model_selection import GridSearchCV
from sklearn.svm import SVC

# Define the parameter grid to search
param_grid = {
    'C': [0.1, 1, 10, 100],  # Regularization parameter
    'kernel': ['linear', 'rbf', 'poly'],  # Kernel type
    'gamma': ['scale', 'auto']  # Kernel coefficient
}

# Create a GridSearchCV object
grid_search = GridSearchCV(SVC(), param_grid, cv=5, scoring='accuracy')

# Fit the GridSearchCV object to the training data
grid_search.fit(x_train, y_train)

# Print the best hyperparameters and the corresponding accuracy score
print("Best Hyperparameters:", grid_search.best_params_)
print("Best Accuracy Score:", grid_search.best_score_)

# Evaluate the model with the best hyperparameters on the test data
best_svm_model = grid_search.best_estimator_
y_pred_test = best_svm_model.predict(x_test)
cf = classification_report(y_pred_test, y_test)
print('Classification Report:\n', cf)
```
