```python
import pandas as pd
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
file_path='sales_data_with_discounts.csv'
data=pd.read_csv(file_path)
```

```python
data.head
```

```python
print("dataset head:")
print(data.head())
print("\ndataset info:")
data.info()
```

```python
numerical_cols=data.select_dtypes(include=[np.number])
descriptive_stats=numerical_cols.describe()
print(descriptive_stats)
```

```python
for col in numerical_cols:
  mean=data[col].mean()
  median=data[col].median()
  mode=data[col].mode()[0]
  std_dev=data[col].std()
  print(f"nstatistics for {col}:\"nMean: {mean}","Median:{median}","Mode:{mode}","Standard Deviation:{std_dev}")
```

```python
for col in numerical_cols:
  plt.figure(figsize=(5,4))
  sns.histplot(data[col],kde=True)
  plt.title(f"histogram for{col}")
  plt.show()
```

```python
for col in numerical_cols:
  plt.figure(figsize=(5,4))
  sns.boxplot(x=data[col])
  plt.title(f"boxplot for{col}")
  plt.show()
```

```python
categorical_cols=data.select_dtypes(include=['object']).columns
for col in categorical_cols:
  plt.figure(figsize=(5,4))
  sns.countplot(y=data[col],order=data[col].value_counts().index)
  plt.title(f"bar chart for{col}")
  plt.show()
```

```python
from sklearn.preprocessing import StandardScaler

scaler =StandardScaler()
standardized_data=pd.DataFrame(scaler.fit_transform(numerical_cols),columns=numerical_cols.columns)
print("standardized data:")
print(standardized_data.head())
```

```python
categorical_cols = data.select_dtypes(include=['object']).columns

encoded_data = pd.get_dummies(data[categorical_cols], drop_first=True)

# Display a portion of the transformed dataset
print(encoded_data.head())
```
